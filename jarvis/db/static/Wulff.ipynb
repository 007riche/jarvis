```python
import urllib.request
import json
from pymatgen.core.structure import Structure
from pymatgen.analysis.wulff import WulffShape
from pymatgen.core.lattice import Lattice

url = 'https://ndownloader.figshare.com/files/14146319' 
urllib.request.urlretrieve(url, 'jff2.json') 

```

```python
f=open('jff2.json','r')
data=json.load(f)
f.close()
```

```python
data[0]
```

```python
Al_fcc = Structure(Lattice.cubic(4.05), ["Al", "Al", "Al", "Al"],
                           [[0.000000, 0.000000, 0.000000],
                           [0.000000, 0.500000, 0.500000],
                           [0.500000, 0.000000, 0.500000],
                           [0.500000, 0.500000, 0.000000]])
for i in data:
    if i['name']=='JLMP-1243':   
        surf_val=(i['surf_val'])
        tmp=i['surf_name']
        surf_name=[]
        for j in tmp:
           surf_name.append([int(m) for m in j.split('_')])
surf_name

wulff_Al = WulffShape(Al_fcc.lattice, surf_name, surf_val).get_plot()
wulff_Al.show()
wulff_Al.show()
```
