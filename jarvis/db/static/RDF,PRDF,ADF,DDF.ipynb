```python
import os
from pymatgen.core.structure import Structure
from jarvis.sklearn.get_desc import get_rdf,get_prdf,ang_dist1,ang_dist2,dihedrals,get_dist_cutoffs,get_structure_data
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
si_path = os.path.join(os.getcwd(), '..','..','vasp','examples','SiOptb88')
s=Structure.from_file(os.path.join(si_path,'MAIN-RELAX-bulk@mp_149','POSCAR'))
```

```python
prdf=get_prdf(s=s,plot_prdf=True)
prdf.show()
```

```python
dist_rdf,bins_rdf,bo_rdf=get_rdf(s=s)
plt.plot(dist_rdf,bins_rdf)
plt.xlabel('r ($\AA$)')
plt.ylabel('RDF')
```

```python
rcut,rcut1,rcut2,rcut_dihed=get_dist_cutoffs(s=s)
print ('rcut_dihed',rcut_dihed)
struct_info=get_structure_data(s=s)
adfa,bins=ang_dist1(struct_info=struct_info,plot=False,rcut=rcut_dihed)
adfb,_=ang_dist2(struct_info=struct_info,plot=False,rcut2=rcut2)
ddf,_=dihedrals(struct_info=struct_info,plot=False,rcut_dihed=rcut_dihed)
```

```python
plt.plot(bins[:-1],adfa)
plt.xlabel('Angle(degree)')
plt.ylabel('Angle distribution-1st')
```

```python
plt.plot(bins[:-1],adfb)
plt.xlabel('Angle(degree)')
plt.ylabel('Angle distribution-2nd')
```

```python
plt.plot(bins[:-1],ddf)
plt.xlabel('Angle(degree)')
plt.ylabel('Dihedral distribution-1st')
```
